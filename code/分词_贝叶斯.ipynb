```python
import jieba
import pandas as pd
import os
import re
```

# 分词

```python
cache='data'
csv_list=['train.csv','test.csv','val.cav']
jieba.load_userdict("data/newwords.txt")
```

```python
#去除数字
def remove_digits(input_str):
    punc = u'0123456789.'
    output_str = re.sub(r'[{}]+'.format(punc), '', input_str)
    return output_str

# 去除停用词
def move_stopwords(sentence_list, stopwords_list):
    # 去停用词
    out_list = []
    for word in sentence_list:
        if word not in stopwords_list:
            if not remove_digits(word):
                continue
            if word != '\t':
                out_list.append(word)
    return out_list
# 创建停用词列表
def get_stopwords_list():
    stopwords = [line.strip() for line in open('data/stopwords.txt',encoding='UTF-8').readlines()]
    stopwords.append(' ')
    return stopwords
```

```python
stopwords = get_stopwords_list()
```

```python
sentences = [x.strip().split('\t') for x in open(os.path.join(cache, csv_list[0]),encoding='utf-8').readlines()]
sentences
```

```python
good = []
bad = []
```

```python
# l=jieba.lcut(sentences[2][1])
i=1
for line in sentences:
    print(i)
    i+=1
    if line[0] == '0':
        sentence_depart = jieba.lcut(line[1].strip())
        sentence_depart = move_stopwords(sentence_depart, stopwords)
        for word in sentence_depart:
            bad.append(word)
    else:
        sentence_depart = jieba.lcut(line[1].strip())
        sentence_depart = move_stopwords(sentence_depart, stopwords)
        for word in sentence_depart:
            good.append(word)
```
